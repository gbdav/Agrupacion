```python
import pandas as pd
```

```python
songs = pd.read_csv("source/SpotifyDailyTop200SongsWithGenres2017-2021/data.csv", sep="#")
songs
```

```python
canPorArtista = songs.groupby(['Artist']).size().reset_index(name='canciones')
canPorArtista.sort_values(by='canciones',ascending=False).head(10)
```

```python
repPorArtista = songs.groupby(['Artist'])['Streams'].sum().reset_index(name='reproducciones')
repPorArtista.sort_values(by='reproducciones',ascending=False).head(10)
```

```python
songs.groupby(['Date'])['Streams'].sum().sort_values(ascending=False).head(10)
```
